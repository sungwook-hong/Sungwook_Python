```python
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
%matplotlib inline

start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2017, 6, 8)
apple_df = web.DataReader("AAPL", 'google', start, end)
#plt.plot(apple_df.index, apple_df.Close, label="AAPL's close")
#plt.grid()
#plt.legend(loc = "best")

apple_df.head(10)
```

```python
fig = plt.figure(figsize=(12,8))

top_axes = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
bottom_axes = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4)
bottom_axes.get_yaxis().get_major_formatter().set_scientific(False)

top_axes.plot(apple_df.index, apple_df['Close'], label='Apple\'s Daily Close')
top_axes.legend(loc='best')
bottom_axes.bar(apple_df.index, apple_df['Volume'], label='Volume')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

#subplot2grid is used when we need to change the location or size of the subplot
```

```python
import matplotlib.finance as matfin

fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)

name_list = []
for day in apple_candle_stick.index:
    name_list.append(day.strftime('%m-%d'))
day_list = range(len(apple_candle_stick))

import matplotlib.ticker as ticker
ax.xaxis.set_major_locator(ticker.FixedLocator(day_list))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(name_list))

matfin.candlestick2_ohlc(ax,apple_candle_stick['Open'],apple_candle_stick['High'],apple_candle_stick['Low'],apple_candle_stick['Close'], width=0.5, colorup='g', colordown='r')

plt.show()


```

```python
import matplotlib.finance as matfin

fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)

day_list=[]
name_list=[]
for i, day in enumerate(apple_candle_stick.index):
    if day.dayofweek == 0:
        day_list.append(i)
        name_list.append(day.strftime('%Y-%m-%d')+'(Mon)')

import matplotlib.ticker as ticker
ax.xaxis.set_major_locator(ticker.FixedLocator(day_list))
ax.xaxis.set_major_formatter(ticker.FixedFormatter(name_list))

matfin.candlestick2_ohlc(ax,apple_candle_stick['Open'],apple_candle_stick['High'],apple_candle_stick['Low'],apple_candle_stick['Close'], width=0.5, colorup='g', colordown='r')
plt.grid()
plt.show()
```
